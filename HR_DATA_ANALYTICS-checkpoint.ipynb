## Q15] Identify Department with the Highest attrition rate (Resigned %)? 

```python
df.head(2)
```

```python
df.groupby('Department') ['Status'].agg(total_emp = 'count', resigned = lambda x: (x == 'Resigned').sum())
```

```python
dept_count = df.groupby('Department') ['Status'].agg(total_emp = 'count', resigned = lambda x: (x == 'Resigned').sum())
dept_count
```

```python
type(dept_count)
```

## Calculate Attrition Rate

```python
dept_count ['attrition_rate_%'] = (dept_count ['resigned'] / dept_count ['total_emp']) * 100
dept_count
```

```python
dept_count.sort_values('attrition_rate_%', ascending = False)
```
